London is one of the most common choices for international students in terms of tertiary education. There are a lot of famous universities located in London, including Imperial College London (ICL), University College London (UCL), King's College London (KCL), London School of Economics (LSE) etc. One of the biggest concerns for international students when moving to a new city would be finding an accommodation. In terms of finding a good place to live, there are many criteria that we should concern, for example, living convenience, rent rate, crime rate etc. Therefore in this project, I intend to explore different neighborhoods of London and find the best area for international students to live in London.

This study will focus only on the **safety**, **rental easiness and expenses** and **the general atmosphere** of the neighborhood. For the safety part, I will make use of the London recorded crime data to analyze.

```python
#Create a function to explore all borough
def getNearbyVenues(names, latitudes, longitudes, radius=1000):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['BoroughName', 
                  'Borough Latitude', 
                  'Borough Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
# top 5 venues for each neighborhood
num_top_venues = 5

for bor in grouped_borough['BoroughName']:
    print("----"+bor+"----")
    temp = grouped_borough[grouped_borough['BoroughName'] == bor].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

```python
# Put into pandas dataframe
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]

num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Borough']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
venues_sorted = pd.DataFrame(columns=columns)
venues_sorted['Borough'] = grouped_borough['BoroughName']

for ind in np.arange(grouped_borough.shape[0]):
    venues_sorted.iloc[ind, 1:] = return_most_common_venues(grouped_borough.iloc[ind, :], num_top_venues)

venues_sorted.head()
```

```python
from sklearn.metrics import silhouette_samples, silhouette_score
from sklearn.cluster import KMeans

def plot(x, y, xlabel, ylabel):
    plt.figure(figsize=(20,10))
    plt.plot(np.arange(2, x), y, 'o-')
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.xticks(np.arange(2, x))
    plt.show()
    
indices = []
scores = []
max_range = 20

london_cluster = grouped_borough.drop('BoroughName', 1)
for kclusters in range(2, max_range) :
    
    # Run k-means clustering
    lct = london_cluster
    kmeans = KMeans(n_clusters = kclusters, init = 'k-means++', random_state = 0).fit_predict(lct)
    
    # Gets the score for the clustering operation performed
    score = silhouette_score(lct, kmeans)
    
    # Appending the index and score to the respective lists
    indices.append(kclusters)
    scores.append(score)
    

plot(max_range, scores, "No. of clusters", "Silhouette Score")
```

### Results in terms of Crime rate and Rent data

```python
df_analysis = df_london_clustering_data[['Borough','Latitude','Longitude', 'CrimeToPop']]
df_analysis = pd.merge(df_analysis, df_london_rent, how='outer', on='Borough')
df_analysis.head()
```

Now we have data with the Crime to population data, the counts of rents data, and the average rental expense data. We then need to process these data in terms of a score between 0 to 1, so that we can compare them. 

Please be noticed that in this project, we assume every category has the same importance. so that we add up the score directly without giving them weight. If someone values safety more than expense, he/she can always add a higher weight to the safety score.

```python
df_analysis['Safety'] = (df_analysis['CrimeToPop']-df_analysis['CrimeToPop'].min())/(df_analysis['CrimeToPop'].max()-df_analysis['CrimeToPop'].min())
df_analysis['Safety'] = abs(df_analysis['Safety'] - 1) 
df_analysis.head()
```

```python
df_analysis['Easy to Rent'] = (df_analysis['Count of rents']-df_analysis['Count of rents'].min())/(df_analysis['Count of rents'].max()-df_analysis['Count of rents'].min())
df_analysis['Easy to Rent'] = abs(df_analysis['Easy to Rent'] - 1) 
df_analysis.head()
```

```python
df_analysis['Rent Expense'] = (df_analysis['Average']-df_analysis['Average'].min())/(df_analysis['Average'].max()-df_analysis['Average'].min())
df_analysis['Rent Expense'] = abs(df_analysis['Rent Expense'] - 1) 
df_analysis.head()
```

```python
df_final = df_analysis[['Borough','Latitude','Longitude','Safety','Easy to Rent','Rent Expense']]
df_final['Total Score'] = df_final['Safety']+df_final['Easy to Rent']+df_final['Rent Expense']
df_final.sort_values(by='Total Score', ascending = False, inplace = True)
df_final.head()
```

```python
df_top5 = df_final[:5]

# make a map of Toronto using folium
map = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough in zip(df_top5['Latitude'], df_top5['Longitude'], df_top5['Borough']):
    label = '{}'.format(borough)
    label2 = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label2,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map)  
    
map
```

However, as we can see from the map, they are quite far away from the City center. The reason behind is guessed to be that for boroughs close to city center, the expense and the crime rate would be higher, which is quite reasonable.
